```python
import numpy as np  
import matplotlib.pylab as plt

m = 100
X = 2 * np.random.rand(100, 1) 
y = 4 + 3 * X + np.random.randn(100,1)
```

```python
plt.plot(X, y, 'r*', label='Dados')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

```python
X_new = np.c_[X, np.ones((m,1))]
```

```python
X_new_T = X_new.T
w, b = np.linalg.inv(X_new_T.dot(X_new)).dot(X_new_T).dot(y)
```

```python
w, b
```

```python
# Predição = Peso * Input + Bias
pred = w[0] * X + b[0]
```

```python
plt.plot(X, y, 'r*', label='Dados')
plt.plot(X, pred, 'b-', label='Predição')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

```python
from sklearn.linear_model  import LinearRegression
ling = LinearRegression()
ling.fit(X.reshape(-1,1), y)
# intercept => Bias
# coef => Peso
ling.intercept_, ling.coef_
```

```python
pred2 = ling.predict(X.reshape(-1,1))
```

```python
plt.plot(X, y, 'r*', label='Dados')
plt.plot(X, pred2, 'b-', label='Predição')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```
